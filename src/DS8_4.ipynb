```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
%matplotlib inline
```

```python
laneDetector = pd.read_table("detector_lane_inventory.tsv",
                     na_values=['-'])
eventsTrain = pd.read_table("events_train_holdout.tsv",
                     na_values=['-'], error_bad_lines=False)
```

```python
from urllib2 import urlopen
from pylab import imshow, imread, show
lon = [38.2,39.8]
lat = [-77.8,-76.2]
scale = 700000

url = urlopen('http://parent.tile.openstreetmap.org/cgi-bin/export?'
                      'bbox={lat1:.2f},{lon1:.2f},{lat2:.2f},{lon2:.2f}&'
                      'scale={scale:d}&format=png'.format(lat1=lat[0],
                      lat2=lat[1],
                      lon1=lon[0],
                      lon2=lon[1],
                      scale=scale))
m = imread(url)
```

```python
accidentData = eventsTrain[eventsTrain['event_subtype']=='accident']
accidentData = accidentData[pd.notnull(accidentData['latitude'])]
accidentData = accidentData[pd.notnull(accidentData['longitude'])]

pointDensity = np.vstack([accidentData['longitude'],accidentData['latitude']])
z = gaussian_kde(pointDensity)(pointDensity)
```

```python
fig, ax = plt.subplots()
ax.imshow(m, extent=lat+lon)
mapx = ax.scatter(accidentData['longitude'], accidentData['latitude'], c=z, s=1, edgecolor='')
fig.colorbar(mapx)
plt.show()
```

```python
def GetNullType(row):
    if pd.isnull(row['default_speed']):
        return 'Speed'
    elif pd.isnull(row['latitude']) or pd.isnull(row['longitude']):
        return 'Location'
    elif pd.isnull(row['bearing']):
        return 'Bearing'
    elif pd.isnull(row['interval']):
        return 'Interval'
    else:
        return None
```

```python
laneDetector['BadData'] = laneDetector.apply(GetNullType, axis=1).dropna()
groupedNulls = laneDetector.groupby(['detector_type','BadData'])
groupedNulls.size().unstack().plot(kind='bar',stacked=True,  color=['red','blue', 'green'], grid=False)
```

```python
### SECOND DATA VISUALIZATION ###
'''
    EXPLANATION:
    This bar graph depicts which type of detectors are faulty with respect to which data. We check for null values in speed,
    location,bearing and interval values. This is following the assumption that a detector records NaN value if it doesn't have
    a proper recording.
    We can deduce that
        -> Microwave type detectors are worst at recording speeds.
        -> Single Loop detectors are worst at recording bearing.
        -> RTMS detectors make most mistakes in recording location.
        -> No detector makes any mistakes to record the interval for this data set.
'''
```
