```python
%pylab inline
import pylab
from sklearn.datasets import fetch_mldata
DATA_PATH = '~/Desktop/Labs/lab04/data'
mnist = fetch_mldata('MNIST original', data_home=DATA_PATH)
```

```python
# Display several images in a row
def show(imgs, n=1):
    fig = pylab.figure()
    for i in xrange(0, n):
        fig.add_subplot(1, n, i+1, xticklabels=[], yticklabels=[])
        if n == 1:
            img = imgs
        else:
            img = imgs[i]
        pylab.imshow(img.reshape(28, 28), cmap="Greys")
```

```python
train = mnist.data[:60000]
test = mnist.data[60000:]
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute').fit(train)
```

```python
differentTestDigits = [0, 1000, 3000, 4000, 5000, 6000, 6500, 7000, 8500, 9500]
for queryImage in differentTestDigits:
    _, result = model.kneighbors(test[queryImage], n_neighbors=4)
    show(test[queryImage])
    show(train[result[0],:], len(result[0]))
```

```python
train = mnist.data[:60000]
train_labels = mnist.target[:60000]

test = mnist.data[60000:]
test_labels = mnist.target[60000:]

# For traning
train_sample = train[::100]
train_sample_labels = train_labels[::100]
# for testing
test_sample = test[::10]
test_sample_labels = test_labels[::10]
```

```python
def getMax(nbConfusionMatrix):
    maxVal = -1
    maxI = 0
    maxJ = 0
    for i in range(0, 10):
        for j in range(0,10):
            if i != j:
                if nbConfusionMatrix[i,j] > maxVal:
                    maxVal = nbConfusionMatrix[i,j]
                    maxI = i
                    maxJ = j
    print maxI, ' is most confused with ', maxJ
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
naiveBayesModel = GaussianNB().fit(train_sample, train_sample_labels)
naiveBayesPred = naiveBayesModel.predict(test_sample)
naiveBayesErrors = [i for i in xrange(0, len(test_sample)) if naiveBayesPred[i] != test_sample_labels[i]]
nbErrRate = float(len(naiveBayesErrors))/len(naiveBayesPred)
print nbErrRate, '\n'
nbConfusionMatrix = confusion_matrix(test_sample_labels, naiveBayesPred)
print nbConfusionMatrix, '\n'
getMax(nbConfusionMatrix)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
svmModel = SVC(kernel='linear').fit(train_sample, train_sample_labels)
svmPred = svmModel.predict(test_sample)
svmErrors = [i for i in xrange(0, len(test_sample)) if svmPred[i] != test_sample_labels[i]]
svmErrRate = float(len(svmErrors))/len(svmPred)
print svmErrRate, '\n'
svmConfusionMatrix = confusion_matrix(test_sample_labels, svmPred)
print svmConfusionMatrix, '\n'
getMax(svmConfusionMatrix)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
lrModel = LogisticRegression().fit(train_sample, train_sample_labels)
lrPred = lrModel.predict(test_sample)
lrErrors = [i for i in xrange(0, len(test_sample)) if lrPred[i] != test_sample_labels[i]]
lrErrRate = float(len(lrErrors))/len(lrPred)
print lrErrRate, '\n'
lrConfusionMatrix = confusion_matrix(test_sample_labels, lrPred)
print lrConfusionMatrix, '\n'
getMax(lrConfusionMatrix)
```
