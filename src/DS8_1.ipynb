```python
import pylab
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from urllib2 import urlopen
from pylab import imshow, imread, show
%matplotlib inline
```

```python
data = pd.read_table("detector_lane_inventory.tsv",
                     na_values=['-'])
data = data[np.invert(np.isnan(data['latitude']))]
data = data[np.invert(np.isnan(data['longitude']))]
```

```python
latitude = data['latitude'].astype(np.float32)
longitude = data['longitude'].astype(np.float32)
print latitude.min(),latitude.max(),longitude.min(),longitude.max()
```

```python
lon = [38.2,39.8]
lat = [-77.8,-76.2]
scale = 700000

url = urlopen('http://parent.tile.openstreetmap.org/cgi-bin/export?'
                      'bbox={lat1:.2f},{lon1:.2f},{lat2:.2f},{lon2:.2f}&'
                      'scale={scale:d}&format=png'.format(lat1=lat[0],
                      lat2=lat[1],
                      lon1=lon[0],
                      lon2=lon[1],
                      scale=scale))
m = imread(url)
```

```python
fig, ax = plt.subplots()
ax.imshow(m, extent=lat+lon)
ax.scatter(longitude, latitude, s=1)
plt.show()
```
