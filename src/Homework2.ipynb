```python
import pylab
import pandas as pd
```

```python
data = pd.read_csv("wc_day6_1_sample.csv",
                     names=['ClientID', 'Date', 'Time', 'URL', 'ResponseCode', 'Size'],
                     na_values=['-'])
```

```python
# Answer 1
groupBy_CodeNDate = data.groupby(['ResponseCode', 'Date'])
answer1 = groupBy_CodeNDate.get_group((404, '30/Apr/1998'))
answer1.shape[0]
```

```python
# Answer 2
may1_Data = data[data['Date']=='01/May/1998']
may1_Data['DateTime'] = pd.to_datetime(may1_Data.apply(lambda row: row['Date'] + ' ' + row['Time'], axis=1))
hour_grouped = may1_Data.groupby(lambda x: may1_Data['DateTime'][x].hour)
```

```python
%matplotlib inline
ax = hour_grouped['ClientID'].nunique().plot(kind='bar')
```

```python
# Answer 3
# The number of unique users and the size of the trafic follow similar trends with only a few outliers(15-20 hour)
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
x = hour_grouped.size().index

ax1.plot(x, hour_grouped['ClientID'].nunique(), 'g-')
ax2.plot(x, hour_grouped['Size'].sum(), 'r-')

ax1.set_xlabel('Hour')
ax1.set_ylabel('No of unique users', color='g')
ax2.set_ylabel('Amount of traffic', color='r')
```

```python
# Answer 4
clients = data['ClientID'].unique()[0:100]
clientVisitData = data[data['ClientID'].isin(clients)]
clientVisitData = clientVisitData.drop_duplicates(subset='ClientID')
clientVisitData['DateTime'] = pd.to_datetime(clientVisitData.apply(lambda row: row['Date'] + ' ' + row['Time'], axis=1))
clientHour_grouped = clientVisitData.groupby(lambda x: clientVisitData['DateTime'][x].hour)
```

```python
dictToList = clientHour_grouped['ClientID'].unique()
result = {}
for x in range(0, 23):
    if x in dictToList:
        for id in dictToList[x]:
            result[id] = x
resultDF = pd.DataFrame(result.items(), columns=['ClientID', 'Hours'])
resultDF.plot(kind='scatter', x='Hours', y='ClientID')
```
